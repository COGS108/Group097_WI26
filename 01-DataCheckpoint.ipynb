## Research Question

What is the relationship between screen time before bed, physical activity levels, sleep quality, and academic performance among US college students aged 18 to 25? Specifically, we want to examine whether screen use close to bedtime is associated with poorer sleep quality (measured using the Pittsburgh Sleep Quality Index), and whether that in turn is related to lower academic performance such as exam scores, homework scores, lecture attendance, and overall GPA. We also want to explore whether physical activity acts as a moderating variable in this relationship, meaning does regular exercise reduce or buffer the negative effects that screen time before bed has on sleep and academics? We will also control for demographic and institutional factors such as age, year in school (freshman through senior), and type of institution (community college, public university, private university, etc.) to make sure any relationships we find aren't being driven by those differences. The goal is to figure out not just whether screen time before bed hurts sleep and grades, but whether US college students who are more physically active are less affected by it compared to students who aren't.


## Background and Prior Work

Living in an era where technology is constantly evolving, it has impacted our lives immensely. We use smartphones and screens for almost every aspect of our student lives, from attending virtual lectures to unwinding with social media after a long day of classes. Research has shown that this constant connectivity is fundamentally shifting sleep patterns among students and, as a result, their ability to perform academically [1, 2]. Because technology is so integrated into our generation's lifestyle, we often don't realize how much our late-night scrolling might be the culprit behind our morning brain fog.

This topic is relevant to our team as we've noticed how our own academic motivation seems to dip when our screen time spikes. Studies have found that excessive screen exposure disrupts the delicate balance required for restorative rest and cognitive functioning [1, 2]. The reason this happens is because screens emit blue light which suppresses melatonin production, the hormone responsible for signaling to your body that it's time to sleep [7, 8]. On top of that, social media use triggers cognitive arousal, where the brain is constantly processing new content, reacting to posts, and receiving small dopamine hits from notifications, making it much harder to wind down before bed [2, 8]. A systematic review of blue light research in young adults confirmed that evening blue light exposure delays circadian phase and increases sleep onset latency, leading to later bedtimes and shorter sleep duration [7]. Research has also found a direct link between these digital habits and lower academic outcomes, with students who have higher screen time consistently performing worse on exams and attending fewer classes [3, 6].

To summarize the work done previously by other researchers, we found that high screen time is consistently correlated with delayed sleep onset and reduced total sleep duration [1]. Previous cross-sectional analyses have shown that for young adults and children, excessive screen use is a primary predictor of poor sleep quality, often due to the blue light exposure and cognitive arousal mechanisms described above [2, 7]. We find it particularly interesting that sleep appears to act as a "mediator." This means that while screen use impacts grades, it might be doing so because it ruins sleep first [5].

Furthermore, previous studies have demonstrated that when screen time is limited, there are measurable improvements in sleep duration and school grades [4]. Other predictive modeling has shown that as screen time increases, the probability of achieving high academic marks decreases [6].

On the other side of things, physical activity has been consistently shown to improve sleep quality across various populations. A systematic review found that regular exercise reduces sleep onset latency, increases total sleep duration, and improves overall sleep quality [10]. A study of US adolescents aged 16 to 19 found that physical activity, screen time, and sleep are all closely related, with higher physical activity levels associated with better sleep outcomes even among students with high screen time [9]. What really caught our attention is a study on college-aged adults that found physical activity and sleep actually moderate the relationship between stress and screen time, where in highly active individuals screen time was essentially unrelated to stress levels [11]. This suggests that exercise might serve as a protective factor that buffers the negative effects of screen time on sleep and well-being.

However, most existing research looks at screen time, sleep, and physical activity as separate pieces rather than examining how they interact with each other. It's also worth noting that factors like age, year in school, and institution type could all influence screen time habits, physical activity levels, and academic performance differently. A freshman at a large public university might have a very different lifestyle and schedule than a senior at a small private school, so we plan to control for these demographic and institutional variables in our analysis. We want to bring all of these pieces together and see whether physical activity can actually reduce the damage that screen time before bed does to sleep quality and academic performance among US college students.

## Hypothesis


We hypothesize that higher screen time before bed will be associated with poorer sleep quality as measured by the Pittsburgh Sleep Quality Index (PSQI) among US college students aged 18 to 25, and that poorer sleep quality will in turn be associated with worse academic performance including lower exam scores, reduced lecture attendance, lower homework scores, and a lower overall GPA. We also predict that sleep quality will act as a mediator in this relationship, meaning screen time before bed doesn't directly hurt grades but rather does so by disrupting sleep first through the blue light and cognitive arousal mechanisms discussed in our background section. More importantly, we hypothesize that physical activity will act as a moderating variable. Specifically, we predict that US college students who are more physically active will show a weaker negative association between screen time before bed and sleep quality compared to students who are less active. In other words, we think exercise can partially buffer the negative effects of screen time on sleep because physical activity has been independently shown to improve sleep onset latency and sleep duration. For college students who don't exercise much, we expect the screen time before bed to sleep quality relationship to be much stronger. We expect these predictions to hold even after controlling for age, year in school, and institution type, since these demographic and institutional factors could independently influence both screen time habits and academic outcomes. As discussed in our background, research on college-aged adults has shown that in highly active individuals screen time was essentially unrelated to stress levels, suggesting that physical activity provides some kind of protective effect against the downsides of screen use. Prior research on US adolescents has also found that physical activity, screen time, and sleep quality are all closely interrelated, further supporting our predictions.

### Dataset #1: Student Insomnia and Educational Outcomes (Mendeley)

This is our primary dataset for analyzing the relationship between screen time before bed, physical activity, sleep quality, and academic performance. It contains 995 survey responses collected via Google Forms in October-November 2024 from students at Daffodil International University in Bangladesh. The dataset has 16 variables in categorical or ordinal format.

The most important variables for our research question are electronic device usage before sleep (which directly maps to our screen time before bed variable), physical activity frequency, and sleep quality. For academic outcomes it has class attendance, assignment completion, concentration levels, and an overall academic performance measure. It also includes potentially useful control variables like caffeine consumption, stress levels from academic workload, year of study, and gender.

The main concern with this dataset is that it is from a university in Bangladesh, not the US. Cultural differences in screen habits, academic systems, and lifestyle could affect how generalizable the findings are to US college students. The data is also in categorical/ordinal format (like "frequently" vs "rarely" rather than exact hours), which limits us to certain types of statistical tests. However, this is the only real dataset we could find that has device usage before sleep, physical activity, sleep quality, and academic performance all in the same dataset, which makes it essential for testing the core interaction in our research question.

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('data/00-raw/student-insomnia.csv')

# original column names are the full survey questions, renaming for convenience
df1.columns = [
    'timestamp', 'year_of_study', 'gender', 'difficulty_falling_asleep',
    'sleep_hours', 'night_awakenings', 'sleep_quality', 'concentration_difficulty',
    'daytime_fatigue', 'missed_classes', 'assignment_impact', 'device_before_sleep',
    'caffeine_consumption', 'physical_activity', 'academic_stress', 'academic_performance'
]

print(df1.shape)
df1.head()
```

```python
# drop timestamp col (just the form submission time, not useful)
df1 = df1.drop(columns=['timestamp'])

print('Shape:', df1.shape)
print('Duplicate rows:', df1.duplicated().sum())
# already tidy — one row per student, one col per variable
```

```python
# check missing data
print(df1.isnull().sum())
print('\nTotal missing:', df1.isnull().sum().sum())
# no missing values
```

```python
# all categorical so just checking that values make sense
for col in df1.columns:
    print(f'{col}: {list(df1[col].unique())}\n')

# flag any categories with very few responses
print('--- categories with less than 15 responses ---')
for col in df1.columns:
    counts = df1[col].value_counts()
    rare = counts[counts < 15]
    if len(rare) > 0:
        for val, count in rare.items():
            print(f'  {col}, "{val}": {count} responses')
```

```python
# encode categorical responses as ordinal integers
# 0 = lowest, higher = more frequent / better

freq_5_map = {'Never': 0, 'Rarely (1-2 times a week)': 1, 'Sometimes (3-4 times a week)': 2, 
              'Often (5-6 times a week)': 3, 'Every night': 4}

freq_5_alt_map = {'Never': 0, 'Rarely (1-2 times a week)': 1, 'Sometimes (3-4 times a week)': 2, 
                  'Often (5-6 times a week)': 3, 'Every day': 4}

missed_classes_map = {'Never': 0, 'Rarely (1-2 times a month)': 1, 
                      'Sometimes (1-2 times a week)': 2, 'Often (3-4 times a week)': 3, 'Always': 4}

concentration_map = {'Never': 0, 'Rarely': 1, 'Sometimes': 2, 'Often': 3, 'Always': 4}

sleep_quality_map = {'Very poor': 0, 'Poor': 1, 'Average': 2, 'Good': 3, 'Very good': 4}

sleep_hours_map = {'Less than 4 hours': 0, '4-5 hours': 1, '6-7 hours': 2, 
                   '7-8 hours': 3, 'More than 8 hours': 4}

assignment_impact_map = {'No impact': 0, 'Minor impact': 1, 'Moderate impact': 2, 
                         'Major impact': 3, 'Severe impact': 4}

stress_map = {'No stress': 0, 'Low stress': 1, 'High stress': 2, 'Extremely high stress': 3}

performance_map = {'Poor': 0, 'Below Average': 1, 'Average': 2, 'Good': 3, 'Excellent': 4}

year_map = {'First year': 1, 'Second year': 2, 'Third year': 3, 'Graduate student': 4}

gender_map = {'Male': 0, 'Female': 1}

df1['year_of_study'] = df1['year_of_study'].map(year_map)
df1['gender'] = df1['gender'].map(gender_map)
df1['difficulty_falling_asleep'] = df1['difficulty_falling_asleep'].map(freq_5_map)
df1['sleep_hours'] = df1['sleep_hours'].map(sleep_hours_map)
df1['night_awakenings'] = df1['night_awakenings'].map(freq_5_map)
df1['sleep_quality'] = df1['sleep_quality'].map(sleep_quality_map)
df1['concentration_difficulty'] = df1['concentration_difficulty'].map(concentration_map)
df1['daytime_fatigue'] = df1['daytime_fatigue'].map(concentration_map)
df1['missed_classes'] = df1['missed_classes'].map(missed_classes_map)
df1['assignment_impact'] = df1['assignment_impact'].map(assignment_impact_map)
df1['device_before_sleep'] = df1['device_before_sleep'].map(freq_5_map)
df1['caffeine_consumption'] = df1['caffeine_consumption'].map(freq_5_alt_map)
df1['physical_activity'] = df1['physical_activity'].map(freq_5_alt_map)
df1['academic_stress'] = df1['academic_stress'].map(stress_map)
df1['academic_performance'] = df1['academic_performance'].map(performance_map)

# check that mapping didn't create NaNs from unmatched categories
print('NaN after encoding:', df1.isnull().sum().sum())
df1.head()
```

```python
# summary stats on key variables (0=lowest, 4=highest)
key_vars = ['device_before_sleep', 'physical_activity', 'sleep_quality', 
            'sleep_hours', 'academic_performance', 'missed_classes']
df1[key_vars].describe().round(2)
```

```python
df1.to_csv('data/02-processed/student-insomnia-clean.csv', index=False)
print('Saved to data/02-processed/student-insomnia-clean.csv')
print('Final shape:', df1.shape)
```

### Dataset #2: CMU Sleep and GPA (Carnegie Mellon)

This dataset contains objective sleep data collected from 634 first-year college students across three US universities: Carnegie Mellon University, University of Washington, and Notre Dame. Sleep was tracked using Fitbit devices over the course of a semester, which means these are actual measured sleep values rather than self-reported survey responses.

The key variables are:
- **TotalSleepTime**: Average nightly sleep in minutes. For reference, the CDC recommends 7+ hours (420+ minutes) for adults aged 18-25. Values in this dataset range from roughly 250 to 550 minutes.
- **bedtime_mssd**: Mean successive squared difference of bedtime, which measures how inconsistent a student's bedtime is from night to night. Higher values mean more irregular sleep schedules. This is important because sleep regularity has been shown to matter just as much as total sleep duration for cognitive performance.
- **midpoint_sleep**: The midpoint of the sleep period in minutes from midnight. Higher values indicate later sleep timing (e.g., 360 = 6:00 AM midpoint, meaning sleeping roughly 2 AM to 10 AM).
- **daytime_sleep**: Average minutes of daytime sleep (naps). High values could indicate that students aren't getting enough nighttime sleep.
- **frac_nights_with_data**: Fraction of nights where Fitbit data was recorded. Lower values mean more missing nights, which could affect reliability.
- **cum_gpa** and **term_gpa**: Cumulative and term GPA on a 4.0 scale.
- **term_units**: Number of credit units taken that term. Also provided as a z-score (Zterm_units_ZofZ).
- **demo_race**, **demo_gender**, **demo_firstgen**: Demographic variables (coded as 0/1).
- **study** and **cohort**: Identifiers for which university and cohort the student belongs to.

The main limitation of this dataset is that it does not include screen time or physical activity variables, so we can't directly test the full interaction from our research question with this data alone. However, it gives us something the Mendeley dataset doesn't: objective, device-measured sleep data paired with real GPA from US college students. This lets us validate the sleep to academic performance piece of our model with hard data rather than relying only on self-reported survey responses. The Fitbit-based measurements also give us sleep consistency (bedtime_mssd) which is a dimension of sleep quality that survey data usually misses.

```python
df2 = pd.read_csv('data/00-raw/cmu-sleep.csv')
print(df2.shape)
df2.dtypes
```

```python
print('Duplicate rows:', df2.duplicated().sum())
print('Duplicate subject_ids:', df2['subject_id'].duplicated().sum())
# 87 dup subject_ids — same students across multiple cohorts/studies, not actual dupes
df2.head()
```

```python
# some columns have ' ' (space) instead of NaN for missing vals
print('Before fixing:')
print(df2.isnull().sum())

for col in ['demo_race', 'demo_gender', 'demo_firstgen', 'term_units', 'Zterm_units_ZofZ']:
    n_spaces = (df2[col] == ' ').sum()
    if n_spaces > 0:
        print(f'{col} has {n_spaces} hidden missing values (spaces)')

df2.replace(' ', np.nan, inplace=True)
print('\nAfter replacing spaces with NaN:')
print(df2.isnull().sum())
```

```python
# outlier check

print('bedtime_mssd: mean={:.2f}, median={:.2f}, max={:.2f}'.format(
    df2['bedtime_mssd'].mean(), df2['bedtime_mssd'].median(), df2['bedtime_mssd'].max()))
print(f'  {(df2["bedtime_mssd"] > 5).sum()} students with very irregular sleep (mssd > 5)')

print(f'\ndaytime_sleep max: {df2["daytime_sleep"].max():.0f} min')
print(f'  {(df2["daytime_sleep"] > 150).sum()} students averaging 2.5+ hrs of naps/day')

print(f'\n{(df2["frac_nights_with_data"] < 0.5).sum()} students with <50% nights recorded')
print(f'{(df2["frac_nights_with_data"] < 0.3).sum()} students with <30% nights recorded')

print(f'\ncum_gpa: {df2["cum_gpa"].min():.2f} to {df2["cum_gpa"].max():.2f}')
print(f'term_gpa: {df2["term_gpa"].min():.2f} to {df2["term_gpa"].max():.2f}')

# demo_firstgen should be 0/1 binary but there's a 2 in here
print(f'\ndemo_firstgen values:')
print(df2['demo_firstgen'].value_counts(dropna=False))
```

```python
# convert string cols to numeric now that spaces → NaN
for col in ['demo_race', 'demo_gender', 'demo_firstgen', 'term_units', 'Zterm_units_ZofZ']:
    df2[col] = pd.to_numeric(df2[col], errors='coerce')

# demo_firstgen=2 doesn't make sense for a binary var, set to NaN
df2.loc[df2['demo_firstgen'] == 2, 'demo_firstgen'] = np.nan

# keep rows with missing demographics — sleep/GPA data still usable,
# demographics are just control vars

# drop students with <30% fitbit data — not enough nights to be reliable
before = len(df2)
df2 = df2[df2['frac_nights_with_data'] >= 0.3]
print(f'Dropped {before - len(df2)} students with <30% fitbit data')
print(f'Shape: {df2.shape}')
print(f'\nRemaining missing values:')
print(df2.isnull().sum())
```

```python
print(df2[['TotalSleepTime', 'bedtime_mssd', 'midpoint_sleep', 'daytime_sleep', 
            'cum_gpa', 'term_gpa']].describe().round(2))

# GPA and sleep broken down by school
print('\navg term GPA and sleep by school:')
print(df2.groupby('cohort')[['term_gpa', 'TotalSleepTime']].mean().round(2))

# correlations with term_gpa
print('\ncorrelations with term_gpa:')
for col in ['TotalSleepTime', 'bedtime_mssd', 'midpoint_sleep', 'daytime_sleep']:
    r = df2[col].corr(df2['term_gpa'])
    print(f'  {col}: r = {r:.3f}')
```

```python
df2.to_csv('data/02-processed/cmu-sleep-clean.csv', index=False)
print('Saved to data/02-processed/cmu-sleep-clean.csv')
print('Final shape:', df2.shape)
```

## Team Expectations 

* *Team Expectation 1: Communication Protocol- We will use iMessages as our primary method for all daily communication. We expect everyone to acknowledge or respond to messages within 24 hours on weekdays. We will hold weekly virtual meetings to sync on progress, and we agree to notify the group at least 4 hours in advance if we cannot attend a scheduled session.*

* *Team Expectation 2: Tone and Conflict Resolution- We will commit to a "honest and polite" tone. We will use "I" statements (like "I think this visualization is a bit cluttered, what do you think?") to ensure feedback feels constructive rather than personal. If a conflict arises, we will first try to resolve the issue speaking with those involved and if needed come together as a group to hear each other out. But every perspective involved will be heard and respected.*

* *Team Expectation 3: Task Management & Transparency- We will send out to-do lists on messages to track our tasks. Every team member is responsible for what they are assigned to take on and communicating progress with the group. We will not wait until the meeting to communicate that we might be stuck or need help on a task; we will also message the groupchat welcoming any and all questions and updates together. If we feel that we can help a teammate we will step up to do so without taking too much burden.*

* *Team Expectation 4: Shared Contribution & Coding- While we will play to our strengths (as some of us are have large coding backgrounds and others have more research backgrounds), we will still divide up the work. All code must include an appropriate comment and pushed to GitHub via a branch and Pull Request so that at least one other teammate can review and edit it for clarity and accuracy. This way we can keep track and stay organized.* 

* *Team Expectation 5: Decision Making- All decisions will be made by a majority vote. However, for time-sensitive or "urgent" decisions (deadlines within 12 hours), the members present at that time are authorized to make the call to ensure the project moves forward. We will document these decisions in our iMessages and ensure that all members are updated and stay included.*

* *Team Expectation 6: Accountability and Support- If a teammate is struggling with a deadline due to outside responsibilities or technical issues, they must contact the group at least the night before the deadline. We will then reallocate effort as a team to help offer support. However, we acknowledge that if a member consistently fails to meet agreed-upon tasks without communication, we will follow the official policy of sending a written warning before escalating to the professor by Week 7.*
