## Research Question

What is the relationship between screen time before bed, physical activity levels, sleep quality, and academic performance among US college students aged 18 to 25? Specifically, we want to examine whether screen use close to bedtime is associated with poorer sleep quality (measured using the Pittsburgh Sleep Quality Index), and whether that in turn is related to lower academic performance such as exam scores, homework scores, lecture attendance, and overall GPA. We also want to explore whether physical activity acts as a moderating variable in this relationship, meaning does regular exercise reduce or buffer the negative effects that screen time before bed has on sleep and academics? We will also control for demographic and institutional factors such as age, year in school (freshman through senior), and type of institution (community college, public university, private university, etc.) to make sure any relationships we find aren't being driven by those differences. The goal is to figure out not just whether screen time before bed hurts sleep and grades, but whether US college students who are more physically active are less affected by it compared to students who aren't.


## Background and Prior Work

Living in an era where technology is constantly evolving, it has impacted our lives immensely. We use smartphones and screens for almost every aspect of our student lives, from attending virtual lectures to unwinding with social media after a long day of classes. Research has shown that this constant connectivity is fundamentally shifting sleep patterns among students and, as a result, their ability to perform academically [1, 2]. Because technology is so integrated into our generation's lifestyle, we often don't realize how much our late-night scrolling might be the culprit behind our morning brain fog.

This topic is relevant to our team as we've noticed how our own academic motivation seems to dip when our screen time spikes. Studies have found that excessive screen exposure disrupts the delicate balance required for restorative rest and cognitive functioning [1, 2]. The reason this happens is because screens emit blue light which suppresses melatonin production, the hormone responsible for signaling to your body that it's time to sleep [7, 8]. On top of that, social media use triggers cognitive arousal, where the brain is constantly processing new content, reacting to posts, and receiving small dopamine hits from notifications, making it much harder to wind down before bed [2, 8]. A systematic review of blue light research in young adults confirmed that evening blue light exposure delays circadian phase and increases sleep onset latency, leading to later bedtimes and shorter sleep duration [7]. Research has also found a direct link between these digital habits and lower academic outcomes, with students who have higher screen time consistently performing worse on exams and attending fewer classes [3, 6].

To summarize the work done previously by other researchers, we found that high screen time is consistently correlated with delayed sleep onset and reduced total sleep duration [1]. Previous cross-sectional analyses have shown that for young adults and children, excessive screen use is a primary predictor of poor sleep quality, often due to the blue light exposure and cognitive arousal mechanisms described above [2, 7]. We find it particularly interesting that sleep appears to act as a "mediator." This means that while screen use impacts grades, it might be doing so because it ruins sleep first [5].

Furthermore, previous studies have demonstrated that when screen time is limited, there are measurable improvements in sleep duration and school grades [4]. Other predictive modeling has shown that as screen time increases, the probability of achieving high academic marks decreases [6].

On the other side of things, physical activity has been consistently shown to improve sleep quality across various populations. A systematic review found that regular exercise reduces sleep onset latency, increases total sleep duration, and improves overall sleep quality [10]. A study of US adolescents aged 16 to 19 found that physical activity, screen time, and sleep are all closely related, with higher physical activity levels associated with better sleep outcomes even among students with high screen time [9]. What really caught our attention is a study on college-aged adults that found physical activity and sleep actually moderate the relationship between stress and screen time, where in highly active individuals screen time was essentially unrelated to stress levels [11]. This suggests that exercise might serve as a protective factor that buffers the negative effects of screen time on sleep and well-being.

However, most existing research looks at screen time, sleep, and physical activity as separate pieces rather than examining how they interact with each other. It's also worth noting that factors like age, year in school, and institution type could all influence screen time habits, physical activity levels, and academic performance differently. A freshman at a large public university might have a very different lifestyle and schedule than a senior at a small private school, so we plan to control for these demographic and institutional variables in our analysis. We want to bring all of these pieces together and see whether physical activity can actually reduce the damage that screen time before bed does to sleep quality and academic performance among US college students.

## Hypothesis

We also predict that sleep quality will act as a mediator in this relationship, meaning screen time before bed doesn't directly hurt grades but rather does so by disrupting sleep first through the blue light and cognitive arousal mechanisms discussed in our background section. More importantly, we hypothesize that physical activity will act as a moderating variable. Specifically, we predict that US college students who are more physically active will show a weaker negative association between screen time before bed and sleep quality compared to students who are less active. In other words, we think exercise can partially buffer the negative effects of screen time on sleep because physical activity has been independently shown to improve sleep onset latency and sleep duration. For college students who don't exercise much, we expect the screen time before bed to sleep quality relationship to be much stronger. We expect these predictions to hold even after controlling for age, year in school, and institution type, since these demographic and institutional factors could independently influence both screen time habits and academic outcomes. As discussed in our background, research on college-aged adults has shown that in highly active individuals screen time was essentially unrelated to stress levels, suggesting that physical activity provides some kind of protective effect against the downsides of screen use. Prior research on US adolescents has also found that physical activity, screen time, and sleep quality are all closely interrelated, further supporting our predictions.


## Data

### Data overview

We are using three datasets that each cover different parts of our research question. No single dataset has every variable we need perfectly, so we're using multiple real datasets that complement each other. Each dataset will be analyzed to address different aspects of our question, and together they give us a more complete picture of the screen time, physical activity, sleep, and academic performance relationship.

- Dataset #1: Student Insomnia and Educational Outcomes
  - Dataset Name: Student Insomnia and Educational Outcomes Dataset
  - Link: https://data.mendeley.com/datasets/5mvrx4v62z/3
  - Number of observations: 985 (785 with no missing data)
  - Number of variables: 16
  - Relevant variables: Electronic device usage before sleep, physical activity frequency, sleep quality, sleep hours, difficulty falling asleep, night awakenings, class attendance, assignment completion, academic performance, concentration, caffeine consumption, stress levels, year of study, gender
  - Shortcomings: This dataset is from Daffodil International University in Bangladesh, not the US. The variables are in categorical/ordinal format rather than continuous, which may limit certain types of analysis. However, it is the only real dataset we found that has device usage before sleep, physical activity, sleep quality, and academic performance all in the same place.

- Dataset #2: CMU Sleep and GPA Dataset
  - Dataset Name: Nightly Sleep Time and GPA in First-Years
  - Link: https://cmustatistics.github.io/data-repository/psychology/cmu-sleep.html
  - Number of observations: 634
  - Number of variables: 15
  - Relevant variables: term_gpa, cum_gpa, TotalSleepTime, bedtime_mssd (sleep consistency/variability), daytime_sleep, gender, race, demo_firstgen
  - Shortcomings: This dataset does not include screen time or physical activity variables. However, it is real data collected from actual US college freshmen at three universities (CMU, University of Washington, Notre Dame) using Fitbit devices. It gives us objective sleep measurements and real GPA data, which helps validate the sleep to academic performance connection with real US student data.

- Dataset #3: StudentLife Dataset (Dartmouth)
  - Dataset Name: StudentLife
  - Link: https://www.kaggle.com/datasets/dartweichen/student-life
  - Number of observations: 48 students tracked over 10 weeks
  - Number of variables: Sensor data across 10 categories plus surveys and GPA
  - Relevant variables: Physical activity (stationary, walk, run), sleep (bedtime, duration, wake up), phone lock/usage data (proxy for screen time), GPA, stress, mood
  - Shortcomings: The sample size is small at only 48 students from a single university (Dartmouth). However, it contains very rich and detailed sensor data collected passively from students' phones over an entire academic term, including objective measurements of physical activity, sleep, and phone usage alongside real GPA. This makes it useful for validating relationships at a more granular level.

Together, Dataset #1 is our primary dataset since it has all key variables (device use before sleep, physical activity, sleep quality, academic outcomes). Dataset #2 grounds the sleep to GPA relationship in real objective data from US college students. Dataset #3 adds rich sensor-based data on physical activity, phone usage, and sleep with real GPA from US college students, even though the sample is small.
