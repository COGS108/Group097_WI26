## Research Question

What is the relationship between time spent on social media between the hours of 10 PM to 2 AM, and short-term academic performance such as lecture attendance, homework score, and exam scores? Specifically, we want to examine whether screen use within one hour before bedtime, are associated with sleep quality (measured using a standardized sleep scale), and whether these sleep variables are related to both short-term and long-term academic performance (exam scores and GPA). The goal is to determine whether more screen exposure during later hours in the night is linked to poorer sleep and, in turn, lower academic performance.


## Background and Prior Work

Living in an era where technology is constantly evolving, it has impacted our lives immensely. We use smartphones and screens for almost every aspect of our student lives, from attending virtual lectures to unwinding with social media after a long day of classes. However, we believe that the downside of this constant connectivity is that it is fundamentally shifting our sleep patterns and, as a result, our ability to perform academically. Because technology is so integrated into our generation’s lifestyle, we often don't realize how much our late-night scrolling might be the culprit behind our morning brain fog.

This topic is relevant to our team as we’ve noticed how our own academic motivation seems to dip when our screen time spikes. We believe that excessive screen exposure disrupts the delicate balance required for restorative rest and cognitive functioning [1, 2]. We’ve seen firsthand how a night spent on devices leads to a struggle to focus the next day, and we believe there is a direct link between these digital habits and the grades we see on our transcripts [3, 6].

To summarize the work done previously by other researchers, we found that high screen time is consistently correlated with delayed sleep onset and reduced total sleep duration [1]. Previous cross-sectional analyses have shown that for young adults and children, excessive screen use is a primary predictor of poor sleep quality, often due to blue light exposure and cognitive arousal before bed [2]. We find it particularly interesting that sleep appears to act as a "mediator." This means that while screen use impacts grades, it might be doing so because it ruins sleep first [5].

Furthermore, previous studies have demonstrated that when screen time is limited, there are measurable improvements in sleep duration and school grades [4]. Other predictive modeling has shown that as screen time increases, the probability of achieving high academic marks decreases [6]. Because of these findings, we believe that our own research will reveal a similar relationship. Thus, we decided to conduct this research project to see if there is a statistically significant relationship between screen time, sleep quality, and academic achievement within our own target population.

## Hypothesis


We hypothesize that greater time spent using social media applications between 10 PM and 2 AM will lead to poorer sleep quality measured by our standardized sleep scale. Specifically we believe that using social media within 1 hour of going to sleep will have a much bigger effect. We further predict that the poor sleep quality will lead to worse short-term academic performance which includes, lower test scares, less lectures being attended, lower homework scores, and a reduction in overall GPA. We based our hypothesis based on prior research that suggests that phone usage before sleep disrupts sleep patterns, which furthermore affects cognitive functions and academic outcomes. 

## Data

The ideal dataset for this research question would be a dataset that includes detailed information on late night social media use, sleep quality, and academic performance, catered more towards college students. Key variables in this dataset would be, time spent on social media between 10 PM and 2 AM, screentime use within 1 hour of sleep, student GPA, sleep quality scores, lecture attendence, and possibly average exam or homework scores. Other more demographic variables such as age, grade level, region, and school can be useful information. The dataset should include several hundred students in order to account for variation and to ensure there is a good and large representative population. Data for this can be collected through surveys that record students' sleep time, sleep quality, and social media usage before sleep, along with academic information about the students such as GPA or lecture attendance. This data would be stored in a structured database where each row represents a studenta nd each column represents a different variable we are measuring. 

Three data sets:
