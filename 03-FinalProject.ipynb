## Abstract

We investigated whether screen time before bed predicts worse sleep, whether that sleep disruption carries over into academic performance, and whether physical activity buffers any of it. We used three complementary datasets: a 978-student survey with all four constructs, Fitbit-measured sleep paired with GPA from 621 US freshmen, and passively sensed phone and activity data from 30 Dartmouth students with complete records.

Screen time before bed predicted difficulty falling asleep in ordinal logistic regression even after controlling for caffeine, stress, gender, and year of study. A mediation analysis with 5,000 bootstrap samples showed a statistically significant but modest indirect effect through sleep difficulty, accounting for about 14% of screen time's total association with grades. The sleep-to-grades link held across measurement methods: both self-reported sleep and Fitbit-measured sleep duration and bedtime regularity predicted GPA. Model comparison showed that a random forest did not substantially outperform logistic regression, suggesting the relationships are largely linear. Permutation tests validated the key associations without distributional assumptions.

The exercise buffering hypothesis did not hold up. The screen time x physical activity interaction was not significant, and stratified analyses showed similar screen-sleep slopes regardless of activity level. Physical activity predicted academic performance on its own but did not moderate the screen-sleep pathway. All three datasets are cross-sectional, so these findings are associational. The survey sample is from Bangladesh, and the sensor dataset is small (n = 30), so generalizability is limited. Longitudinal data with objective screen time and exercise timing would be needed to test causation and revisit the moderation question with more power.

# Research Question

What is the relationship between screen time before bed, physical activity levels, sleep quality, and academic performance among college students? Specifically, we want to examine whether frequent electronic device use close to bedtime is associated with poorer self-reported sleep quality, shorter sleep duration, and more difficulty falling asleep, and whether those sleep outcomes are in turn related to worse academic performance including lower self-reported grades, more missed classes, and difficulty completing assignments. Beyond that, we want to explore whether physical activity acts as a moderating variable in this relationship, meaning does regular exercise reduce or buffer the negative effects that screen time before bed has on sleep and academics? We will control for demographic factors like gender and year in school to make sure any relationships we find aren't being driven by those differences. The goal is to figure out not just whether screen time before bed hurts sleep and grades, but whether college students who are more physically active are less affected by it compared to students who aren't.


## Background and Prior Work

Living in an era where technology is constantly evolving, it has impacted our lives immensely. Students today use smartphones and screens for almost every aspect of their lives, from attending virtual lectures to unwinding with social media after a long day of classes. Research has shown that this constant connectivity is fundamentally shifting sleep patterns among students and, as a result, their ability to perform academically [1, 2]. Because technology is so integrated into this generation's lifestyle, many students don't realize how much late-night scrolling might be affecting their next-day focus and performance.

Studies have found that excessive screen exposure disrupts the delicate balance required for restorative rest and cognitive functioning [1, 2]. The reason this happens is because screens emit blue light which suppresses melatonin production, the hormone responsible for signaling to your body that it's time to sleep [7, 8]. On top of that, social media use triggers cognitive arousal, where the brain is constantly processing new content, reacting to posts, and receiving small dopamine hits from notifications, making it much harder to wind down before bed [2, 8]. A systematic review of blue light research in young adults confirmed that evening blue light exposure delays circadian phase and increases sleep onset latency, leading to later bedtimes and shorter sleep duration [7]. Research has also found a direct link between these digital habits and lower academic outcomes, with students who have higher screen time consistently performing worse on exams and attending fewer classes [3, 6]. This topic is personally relevant to our team as well since we've noticed similar patterns in our own habits as college students, which is what motivated us to dig deeper into the existing research.

To summarize the work done previously by other researchers, we found that high screen time is consistently correlated with delayed sleep onset and reduced total sleep duration [1]. Previous cross-sectional analyses have shown that for young adults and children, excessive screen use is a primary predictor of poor sleep quality, often due to the blue light exposure and cognitive arousal mechanisms described above [2, 7]. We find it particularly interesting that sleep appears to act as a "mediator." This means that while screen use impacts grades, it might be doing so because it ruins sleep first [5].

Furthermore, previous studies have demonstrated that when screen time is limited, there are measurable improvements in sleep duration and school grades [4]. Other predictive modeling has shown that as screen time increases, the probability of achieving high academic marks decreases [6].

On the other side of things, physical activity has been consistently shown to improve sleep quality across various populations. A systematic review found that regular exercise reduces sleep onset latency, increases total sleep duration, and improves overall sleep quality [10]. A study of US adolescents aged 16 to 19 found that physical activity, screen time, and sleep are all closely related, with higher physical activity levels associated with better sleep outcomes even among students with high screen time [9]. What really caught our attention is a study on college-aged adults that found physical activity and sleep actually moderate the relationship between stress and screen time, where in highly active individuals screen time was essentially unrelated to stress levels [11]. This suggests that exercise might serve as a protective factor that buffers the negative effects of screen time on sleep and well-being.

However, most existing research looks at screen time, sleep, and physical activity as separate pieces rather than examining how they interact with each other. It's also worth noting that factors like gender and year in school could influence screen time habits, physical activity levels, and academic performance differently, so we plan to control for these demographic variables in our analysis. We want to bring all of these pieces together and see whether physical activity can actually reduce the damage that screen time before bed does to sleep quality and academic performance among college students.

# Hypothesis


We hypothesize that students who use electronic devices more frequently before bed will have worse sleep quality, get less sleep overall, and have a harder time falling asleep. We think that these sleep problems will then carry over into their academics, leading to lower grades, more missed classes, and a harder time keeping up with assignments. Basically we don't think screen time before bed directly tanks your grades, but rather it messes up your sleep first and that's what actually hurts your performance. This idea of sleep acting as a mediator is backed up by the research we discussed in our background section, where blue light and cognitive arousal from screens are the main reasons sleep gets disrupted in the first place. The part we're most interested in though is whether physical activity changes this relationship. We predict that students who exercise regularly (like 3 or more times a week) won't be as affected by screen time before bed compared to students who barely exercise or don't exercise at all. So even if two students are both scrolling on their phones before sleeping, the one who works out regularly might still sleep fine while the other one doesn't. We're basing this on research discussed in our background showing that exercise on its own improves how fast you fall asleep and how long you stay asleep, and that in highly active college students screen time basically had no relationship with their stress levels. On top of that, we think that sleep quality is going to matter more for academic performance than just the number of hours someone sleeps. You can get 8 hours of sleep but if you were tossing and turning all night or kept waking up, that's not the same as 8 hours of solid rest. Both of our datasets let us look at this since the insomnia survey has sleep quality ratings and sleep hours as separate variables, and the CMU data has total sleep time alongside GPA. We expect these patterns to show up even when we account for things like gender and year in school since those could affect screen time habits and grades on their own.

### Dataset #2: CMU Sleep and GPA (Carnegie Mellon)

This dataset has objective sleep data from 634 first-year college students at three US universities (Carnegie Mellon, University of Washington, and Notre Dame). Sleep was tracked using Fitbit devices over a semester so these are actually measured sleep values and not self-reported survey answers.

The key variables are:
- **TotalSleepTime**: Average nightly sleep in minutes. The CDC recommends 7+ hours (420+ minutes) for adults aged 18-25. Values in this dataset range from around 250 to 550 minutes.
- **bedtime_mssd**: Mean successive squared difference of bedtime, which basically measures how inconsistent a student's bedtime is from night to night. Higher values mean more irregular sleep schedules. This matters because sleep regularity has been shown to be just as important as total sleep duration for cognitive performance.
- **midpoint_sleep**: The midpoint of the sleep period in minutes from midnight. Higher values mean later sleep timing (e.g., 360 = 6:00 AM midpoint, so sleeping roughly 2 AM to 10 AM).
- **daytime_sleep**: Average minutes of daytime sleep (naps). High values might mean students aren't getting enough sleep at night.
- **frac_nights_with_data**: Fraction of nights where Fitbit data was actually recorded. Lower values mean more missing nights which could make the data less reliable.
- **cum_gpa** and **term_gpa**: Cumulative and term GPA on a 4.0 scale.
- **term_units**: Number of credit units taken that term. Also provided as a z-score (Zterm_units_ZofZ).
- **demo_race**, **demo_gender**, **demo_firstgen**: Demographic variables (coded as 0/1).
- **study** and **cohort**: Identifiers for which university and cohort the student belongs to.

The main limitation here is that it doesn't have screen time or physical activity variables, so we can't test the full interaction from our research question with just this data. However, it actually gives us something the Mendeley dataset doesn't which is objective, device-measured sleep data paired with real GPA from US college students. So we can use it to check whether the sleep to academic performance connection holds up with real measured data instead of only relying on self-reported surveys. The Fitbit measurements also give us sleep consistency (bedtime_mssd) which is something that survey data doesn't usually capture.

### Dataset #3: StudentLife (Dartmouth)

This dataset has passively collected sensor data from 48 Dartmouth students' phones over a 10-week academic term. Unlike our other datasets, the measurements here are completely objective since they come from phone sensors rather than surveys. We extracted the following relevant data from the full dataset:

- **sensing/activity/**: Accelerometer-based activity classification (0=stationary, 1=walking, 2=running) logged every few seconds. We aggregate this per student to get the fraction of time spent in each activity state as a measure of physical activity.
- **sensing/phonelock/**: Phone lock and unlock timestamps. The total duration the phone is unlocked serves as a proxy for screen time.
- **education/grades.csv**: Cumulative GPA and term GPA for each student.
- **survey/psqi.csv**: Pittsburgh Sleep Quality Index survey taken pre and post term, giving us a validated sleep quality measure.
- **survey/PerceivedStressScale.csv**: Perceived stress survey taken pre and post term.
- **EMA/response/Exercise/**: Self-reported exercise check-ins throughout the term.

The main limitation is the small sample size (only 31 students have GPA data), but the objective sensor measurements let us validate patterns from Dataset #1's self-reported data.

```python
import os

grades = pd.read_csv('data/00-raw/student-life/education/grades.csv')
grades.columns = grades.columns.str.strip()
grades.rename(columns={'gpa all': 'gpa_cumulative', 'gpa 13s': 'gpa_term', 'cs 65': 'gpa_cs65'}, inplace=True)
grades['uid'] = grades['uid'].str.strip()

print(f'{len(grades)} students with grades')
grades.head()
```

```python
# activity codes: 0=stationary, 1=walking, 2=running
act_dir = 'data/00-raw/student-life/sensing/activity/'
act_rows = []

for f in os.listdir(act_dir):
    if not f.endswith('.csv'):
        continue
    uid = f.replace('activity_', '').replace('.csv', '')
    tmp = pd.read_csv(os.path.join(act_dir, f), header=None, names=['timestamp', 'activity'], dtype=str)
    tmp = tmp[tmp['timestamp'] != 'timestamp']  # some files have a header row
    tmp['activity'] = pd.to_numeric(tmp['activity'].str.strip(), errors='coerce')
    tmp = tmp.dropna(subset=['activity'])
    
    if len(tmp) == 0:
        continue
    
    counts = tmp['activity'].value_counts()
    n = len(tmp)
    act_rows.append({
        'uid': uid,
        'frac_stationary': counts.get(0, 0) / n,
        'frac_walking': counts.get(1, 0) / n,
        'frac_running': counts.get(2, 0) / n,
        'n_readings': n
    })

act_df = pd.DataFrame(act_rows)
act_df['frac_active'] = act_df['frac_walking'] + act_df['frac_running']

print(f'{len(act_df)} students with activity data')
act_df[['uid', 'frac_stationary', 'frac_active']].head()
```

```python
# phone unlock duration as a proxy for screen time
lock_dir = 'data/00-raw/student-life/sensing/phonelock/'
lock_rows = []

for f in os.listdir(lock_dir):
    if not f.endswith('.csv'):
        continue
    uid = f.replace('phonelock_', '').replace('.csv', '')
    tmp = pd.read_csv(os.path.join(lock_dir, f), header=None, names=['start', 'end'])
    tmp = tmp[tmp['start'] != 'start']
    tmp['start'] = pd.to_numeric(tmp['start'], errors='coerce')
    tmp['end'] = pd.to_numeric(tmp['end'], errors='coerce')
    tmp = tmp.dropna()
    
    tmp['dur_min'] = (tmp['end'] - tmp['start']) / 60.0
    tmp = tmp[(tmp['dur_min'] > 0) & (tmp['dur_min'] < 240)]  # drop bad sessions
    
    if len(tmp) == 0:
        continue
    
    days = max((tmp['end'].max() - tmp['start'].min()) / 86400, 1)
    lock_rows.append({
        'uid': uid,
        'daily_screen_min': tmp['dur_min'].sum() / days,
        'avg_session_min': tmp['dur_min'].mean(),
        'n_unlocks': len(tmp)
    })

lock_df = pd.DataFrame(lock_rows)
print(f'{len(lock_df)} students with phone data')
lock_df.describe().round(2)
```

```python
psqi = pd.read_csv('data/00-raw/student-life/survey/psqi.csv')
psqi.columns = psqi.columns.str.strip()
psqi['uid'] = psqi['uid'].str.strip()

# grab the overall sleep quality rating column
qual_col = [c for c in psqi.columns if 'rate your sleep quality' in c.lower()][0]
psqi_pre = psqi[psqi['type'] == 'pre'][['uid', qual_col]].copy()
psqi_pre.columns = ['uid', 'sleep_quality']

print(f'{len(psqi_pre)} students with pre-term sleep quality')
psqi_pre['sleep_quality'].value_counts()
```

```python
df3 = grades[['uid', 'gpa_cumulative', 'gpa_term']].copy()
df3 = df3.merge(act_df[['uid', 'frac_active', 'frac_stationary']], on='uid', how='left')
df3 = df3.merge(lock_df[['uid', 'daily_screen_min', 'avg_session_min']], on='uid', how='left')
df3 = df3.merge(psqi_pre, on='uid', how='left')

print(f'Merged: {df3.shape}')
print(df3.isnull().sum())
df3.head(10)
```

```python
print(df3[['gpa_cumulative', 'gpa_term', 'frac_active', 'daily_screen_min']].describe().round(3))

print('\nCorrelations with gpa_term:')
for col in ['frac_active', 'frac_stationary', 'daily_screen_min', 'avg_session_min']:
    valid = df3.dropna(subset=[col, 'gpa_term'])
    if len(valid) > 2:
        r = valid[col].corr(valid['gpa_term'])
        print(f'  {col}: r = {r:.3f} (n={len(valid)})')
```

## Results

### Exploratory Data Analysis

Our research question asks whether screen time before bed is associated with worse sleep outcomes, whether those sleep outcomes relate to lower academic performance, and whether physical activity moderates any of those effects. To explore this, we split the EDA into three sections, each using a different dataset to address a different part of the question.

We start with the survey dataset because it's the only source that contains all four constructs (screen time, physical activity, sleep, and academics) in one place, which lets us examine the full chain of relationships and test for moderation by activity level. The limitation is that everything is self-reported, so we follow up with the CMU Fitbit data to check whether the sleep-to-GPA relationship still holds when sleep is objectively measured rather than recalled on a survey. Finally, we use the Dartmouth StudentLife sensor data to get objective screen-time and activity proxies, which helps us triangulate the screen and activity side of things. That sample is small (n = 30), so we treat those results as supporting evidence rather than standalone confirmation.

Across all three sections we use Spearman correlations since most of our variables are ordinal or non-normally distributed, and we focus on the direction and consistency of associations across sources rather than making causal claims. To keep this section reproducible and fast to run, we load fully processed files from `data/02-processed` instead of rerunning the wrangling steps above.

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, pearsonr

from modules.eda_helpers import read_csv_first, plot_likert_counts, spearman_matrix, annotated_regplot, ordinal_codes

sns.set_theme(style="whitegrid", context="talk")

df1, p1 = read_csv_first([
    "data/02-processed/student-insomnia-clean.csv",
    "data/02-processed/student_insomnia_processed.csv"
])
df2, p2 = read_csv_first([
    "data/02-processed/cmu-sleep-clean.csv",
    "data/02-processed/cmu_sleep_processed.csv"
])
df3, p3 = read_csv_first([
    "data/02-processed/student-life-clean.csv",
    "data/02-processed/studentlife_processed.csv"
])

print("Loaded:")
print("df1:", p1, df1.shape)
print("df2:", p2, df2.shape)
print("df3:", p3, df3.shape)
```

#### Section 1: Survey Dataset (Screen Time, Activity, Sleep, and Academics)

We start with the survey dataset because it contains all core constructs in one place and lets us examine the complete relationship implied by our research question.

```python
vars_a = ["device_before_sleep", "physical_activity", "sleep_quality", "academic_performance"]
fig = plot_likert_counts(df1, vars_a, "Survey Dataset: Core Variable Distributions")
plt.show()
```

Most of these variables are ordinal and concentrated around the middle to high response categories, which is why we use rank-based association checks like Spearman instead of assuming normal continuous structure. The dataset is entirely self-reported, so recall and social desirability bias could be affecting the responses.

```python
cols_heat = [
    "device_before_sleep", "physical_activity", "sleep_quality",
    "difficulty_falling_asleep", "sleep_hours", "academic_performance"
]
corr = spearman_matrix(df1, cols_heat)

plt.figure(figsize=(9, 7))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0, square=True)
plt.title("Survey Dataset: Spearman Correlation Matrix")
plt.tight_layout()
plt.show()
```

The correlation matrix shows the direction and relative strength of our key pathways.

**Key findings:** The strongest association is between device use before sleep and physical activity (ρ = 0.41), which suggests these behaviors tend to go together, possibly because more engaged or busy students do more of both. Device use has a moderate positive link with difficulty falling asleep (ρ = 0.34) and a negative association with academic performance (ρ = -0.26). Difficulty falling asleep and sleep hours both correlate negatively with academic performance (ρ = -0.23 and ρ = -0.17, respectively). Self-reported sleep quality shows only weak correlations with the other variables (all |ρ| < 0.13), which could reflect limitations in how respondents interpreted the ordinal scale. The matrix supports directional links from screen use and sleep difficulties to lower academic performance, while physical activity appears positively tied to sleep quality (ρ = 0.12, p < 0.001) but also to device use, which we investigate in the moderation analysis below.

```python
if set(["device_before_sleep", "sleep_quality", "physical_activity"]).issubset(df1.columns):
    tmp = df1[["device_before_sleep", "sleep_quality", "physical_activity"]].dropna().copy()

    for col in ["device_before_sleep", "sleep_quality", "physical_activity"]:
        if not pd.api.types.is_numeric_dtype(tmp[col]):
            tmp[col] = ordinal_codes(tmp[col], col)

    tmp["activity_group"] = np.where(
        tmp["physical_activity"] >= tmp["physical_activity"].median(),
        "Higher activity",
        "Lower activity"
    )

    plt.figure(figsize=(10, 6))
    ax = plt.gca()

    # Stacked annotations on the LEFT
    annotated_regplot(
        tmp[tmp["activity_group"] == "Lower activity"],
        "device_before_sleep",
        "sleep_quality",
        ax,
        "Lower activity",
        text_x=0.03,
        text_y=0.97,
        text_ha="left",
    )
    annotated_regplot(
        tmp[tmp["activity_group"] == "Higher activity"],
        "device_before_sleep",
        "sleep_quality",
        ax,
        "Higher activity",
        text_x=0.03,
        text_y=0.89,
        text_ha="left",
    )

    ax.legend()
    ax.set_title("Survey Dataset: Device Use vs Sleep Quality by Activity Group")
    ax.set_xlabel("Device Use Before Sleep")
    ax.set_ylabel("Sleep Quality")
    plt.tight_layout()
    plt.show()
```

The stratified regression shows that the association between device use and sleep quality is weak in both activity groups: nearly zero for the lower activity group (ρ = 0.01, p = .94, n = 167) and only marginally positive for the higher activity group (ρ = 0.06, p = .07, n = 811). No strong evidence that physical activity buffers the screen-sleep quality link in this dataset. The groups are also imbalanced since most students reported higher activity levels, and sleep quality itself showed weak correlations across the board. If a buffering effect exists, it might operate through other sleep outcomes like difficulty falling asleep rather than the self-reported quality rating.

#### Section 2: CMU Fitbit Dataset (Objective Sleep vs GPA)

Next, we test whether sleep-academic relationships persist in objectively measured U.S. data using Fitbit-derived sleep features and GPA.

```python
cols_b = [c for c in ["TotalSleepTime", "bedtime_mssd", "midpoint_sleep", "daytime_sleep", "term_gpa"] if c in df2.columns]
d2 = df2[cols_b].dropna().copy()

# Plot-only cap to reduce leverage from extreme bedtime irregularity values
mssd_cap = d2["bedtime_mssd"].quantile(0.99)
d2_plot = d2[d2["bedtime_mssd"] <= mssd_cap].copy()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
annotated_regplot(d2_plot, "TotalSleepTime", "term_gpa", axes[0])
axes[0].set_title("CMU: Total Sleep Time vs Term GPA")
axes[0].set_xlabel("Total Sleep Time (minutes)")
axes[0].set_ylabel("Term GPA")

annotated_regplot(d2_plot, "bedtime_mssd", "term_gpa", axes[1])
axes[1].set_title("CMU: Bedtime Irregularity vs Term GPA")
axes[1].set_xlabel("Bedtime MSSD (higher = less regular)")
axes[1].set_ylabel("Term GPA")

plt.tight_layout()
plt.show()

print(f"Plotting cap applied for bedtime_mssd at 99th percentile: {mssd_cap:.2f}")
print(f"Points shown in plots: {len(d2_plot)} / {len(d2)}")

print("Correlations against term_gpa (full data, no cap):")
for col in ["TotalSleepTime", "bedtime_mssd", "midpoint_sleep", "daytime_sleep"]:
    if col in d2.columns:
        s = d2[[col, "term_gpa"]].dropna()
        sr, sp = spearmanr(s[col], s["term_gpa"])
        pr, pp = pearsonr(s[col], s["term_gpa"])
        print(f"{col}: Spearman r={sr:.3f}, p={sp:.4f} | Pearson r={pr:.3f}, p={pp:.4f}, n={len(s)}")
```

The CMU Fitbit results back up the sleep-academic link with objectively measured data. Students who slept more had slightly higher GPAs (ρ = 0.15, p < .001), while those with more irregular bedtimes scored lower (ρ = -0.21, p < .001). Later sleep midpoints (ρ = -0.16, p < .001) and more daytime sleep (ρ = -0.11, p = .006) were also associated with lower GPAs. These directions are consistent with what the survey showed, which makes it harder to attribute the pattern to self-report bias alone. The effect sizes are modest, but that's expected given how many factors influence GPA. The main limitation is that this dataset has no screen time or physical activity variables, so it can only validate the sleep-academic pathway.

### Data Analysis

The EDA established directional associations across all three datasets. Here we formalize those with regression models, mediation and moderation tests, model comparison, and permutation-based validation. We start with ordinal regression on the survey data to isolate screen time's effect on sleep, then test whether sleep predicts academic outcomes in both the survey and CMU data. From there we run a mediation analysis and an interaction model to test the two main hypotheses directly. We close with a model comparison, permutation tests, and a cross-dataset predictor ranking to stress-test the findings.

#### Analysis 1: Screen Time Predicting Sleep Outcomes (Survey)

We fit ordinal logistic regressions to see whether screen use before bed predicts difficulty falling asleep and sleep quality after controlling for gender, year of study, caffeine intake, and stress. Given the EDA correlations, we expect screen time to be a strong predictor of sleep difficulty but weaker for sleep quality, which showed surprisingly flat associations across the board.

```python
from statsmodels.miscmodels.ordinal_model import OrderedModel
import statsmodels.api as sm

controls = ['gender', 'year_of_study', 'caffeine_consumption', 'academic_stress']
predictor = 'device_before_sleep'

results_a1 = {}
for outcome in ['difficulty_falling_asleep', 'sleep_quality']:
    X = df1[[predictor] + controls].dropna()
    y = df1.loc[X.index, outcome]
    
    model = OrderedModel(y, X, distr='logit')
    res = model.fit(method='bfgs', disp=False)
    results_a1[outcome] = res
    
    print(f'\n--- {outcome} ---')
    print(res.summary().tables[1])
```

Screen time before bed strongly predicts difficulty falling asleep but has no detectable relationship with sleep quality. The split is telling: device use seems to disrupt the process of falling asleep specifically rather than degrading how students rate their sleep once they're out. Caffeine is an even stronger predictor of sleep difficulty, and both effects hold after controlling for gender, year of study, and stress. The fact that sleep quality doesn't move with screen time suggests that students may not perceive the damage, even as they take longer to fall asleep.

#### Analysis 2: Sleep Predicting Academic Performance (Survey + CMU)

If screen time hurts grades through sleep, then sleep variables should independently predict academic outcomes. We test this with ordinal logistic regression on the survey data and OLS on the CMU Fitbit data. Running the same question on self-reported and device-measured sleep lets us check whether the link holds up outside of a single survey instrument.

```python
# Survey: sleep variables predicting academic performance
sleep_vars = ['sleep_quality', 'sleep_hours', 'difficulty_falling_asleep']
X_surv = df1[sleep_vars + controls].dropna()
y_surv = df1.loc[X_surv.index, 'academic_performance']

model_surv = OrderedModel(y_surv, X_surv, distr='logit')
res_surv = model_surv.fit(method='bfgs', disp=False)
print('--- Survey: Sleep -> Academic Performance (Ordinal Logit) ---')
print(res_surv.summary().tables[1])

# CMU: sleep variables predicting term GPA (OLS)
sleep_preds_cmu = ['TotalSleepTime', 'bedtime_mssd', 'midpoint_sleep', 'daytime_sleep']
cmu_cols = sleep_preds_cmu + ['term_gpa', 'demo_gender']
d2 = df2[cmu_cols].dropna()

X_cmu = sm.add_constant(d2[sleep_preds_cmu + ['demo_gender']])
y_cmu = d2['term_gpa']
ols_cmu = sm.OLS(y_cmu, X_cmu).fit()

print('\n--- CMU: Sleep -> Term GPA (OLS) ---')
print(ols_cmu.summary().tables[1])
print(f'\nR-squared = {ols_cmu.rsquared:.3f}, Adj R-squared = {ols_cmu.rsquared_adj:.3f}')
```

Difficulty falling asleep is the only sleep variable that significantly predicts grades in the survey, with sleep duration marginal and subjective sleep quality not significant at all. Caffeine and stress are both stronger predictors of academic performance than any sleep variable, likely because they capture broader behavioral patterns that affect grades through multiple channels, not just sleep.

The CMU results paint a similar picture with device-measured data. When students fall asleep matters more than how long they sleep: later sleep timing and more daytime napping are both tied to lower GPAs, while total sleep duration has only a small positive effect. The model explains about 8% of GPA variance, modest but reasonable given how many factors beyond sleep determine grades.

#### Analysis 3: Mediation

Mediation analysis lets us test whether screen time affects grades indirectly, through disrupted sleep. If sleep difficulty is the mechanism, the direct screen-to-grades path should shrink once we control for it. We use Baron & Kenny's (1986) three-step approach with 5,000 bootstrap samples to estimate the indirect effect. Based on the EDA, we expect a partial mediation where screen time still has some direct association with grades but a meaningful chunk runs through sleep difficulty.

```python
med_df = df1[['device_before_sleep', 'difficulty_falling_asleep', 'academic_performance'] + controls].dropna()

# Path a: screen -> mediator
X_a = sm.add_constant(med_df[['device_before_sleep'] + controls])
path_a = sm.OLS(med_df['difficulty_falling_asleep'], X_a).fit()

# Path b + c': mediator + screen -> outcome
X_bc = sm.add_constant(med_df[['device_before_sleep', 'difficulty_falling_asleep'] + controls])
path_bc = sm.OLS(med_df['academic_performance'], X_bc).fit()

# Path c (total): screen -> outcome
X_c = sm.add_constant(med_df[['device_before_sleep'] + controls])
path_c = sm.OLS(med_df['academic_performance'], X_c).fit()

a = path_a.params['device_before_sleep']
b = path_bc.params['difficulty_falling_asleep']
c = path_c.params['device_before_sleep']
c_prime = path_bc.params['device_before_sleep']
indirect = a * b

print(f'Path a (screen -> sleep difficulty):  {a:.4f}, p = {path_a.pvalues["device_before_sleep"]:.4g}')
print(f'Path b (sleep difficulty -> grades):  {b:.4f}, p = {path_bc.pvalues["difficulty_falling_asleep"]:.4g}')
print(f'Path c  (total effect):              {c:.4f}, p = {path_c.pvalues["device_before_sleep"]:.4g}')
print(f"Path c' (direct, controlling M):     {c_prime:.4f}, p = {path_bc.pvalues['device_before_sleep']:.4g}")
print(f'Indirect effect (a*b):               {indirect:.4f}')
print(f'Proportion mediated:                 {indirect/c:.1%}')

# Bootstrap CI for indirect effect
np.random.seed(42)
n_boot = 5000
boot_indirect = np.empty(n_boot)
for i in range(n_boot):
    idx = np.random.choice(len(med_df), len(med_df), replace=True)
    bdf = med_df.iloc[idx]
    Xa = sm.add_constant(bdf[['device_before_sleep'] + controls])
    a_b = sm.OLS(bdf['difficulty_falling_asleep'], Xa).fit(disp=False).params['device_before_sleep']
    Xbc = sm.add_constant(bdf[['device_before_sleep', 'difficulty_falling_asleep'] + controls])
    b_b = sm.OLS(bdf['academic_performance'], Xbc).fit(disp=False).params['difficulty_falling_asleep']
    boot_indirect[i] = a_b * b_b

ci_lo, ci_hi = np.percentile(boot_indirect, [2.5, 97.5])
print(f'\nBootstrap 95% CI for indirect effect: [{ci_lo:.4f}, {ci_hi:.4f}]')
print('(CI excludes 0 = significant mediation)')
```

Sleep difficulty partially mediates the screen time to grades relationship, accounting for about 13.5% of the total effect. The bootstrap confidence interval for the indirect effect excludes zero, so the mediation is statistically significant. The direct path from screen time to grades remains significant even after controlling for sleep difficulty, which means sleep is one mechanism but not the only one. Screen time probably drags on grades through reduced study time, cognitive fatigue, or other pathways that this model can't isolate.

#### Analysis 4: Physical Activity as a Moderator

We add a screen time x physical activity interaction term to the sleep difficulty model. If exercise buffers the damage from late-night screen use, the interaction should be negative, meaning the screen-sleep slope flattens out for more active students. The EDA stratified plot didn't show a clear difference between activity groups, so we're not confident the interaction will be significant, but it's the central hypothesis and worth testing formally.

```python
# Interaction model: screen * activity -> sleep difficulty
mod_df = df1[['device_before_sleep', 'physical_activity', 'difficulty_falling_asleep'] + controls].dropna()
mod_df['screen_x_activity'] = mod_df['device_before_sleep'] * mod_df['physical_activity']

X_mod = sm.add_constant(mod_df[['device_before_sleep', 'physical_activity', 'screen_x_activity'] + controls])
y_mod = mod_df['difficulty_falling_asleep']
res_mod = sm.OLS(y_mod, X_mod).fit()

print(res_mod.summary().tables[1])
print(f'\nR-squared = {res_mod.rsquared:.3f}')

# Visualize: predicted sleep difficulty at low/high activity across screen time levels
fig, ax = plt.subplots(figsize=(8, 5))
screen_range = np.arange(0, 5)
for act_lvl, label, color in [(1, 'Low activity (1)', '#E15759'), (3, 'High activity (3)', '#4E79A7')]:
    pred = (res_mod.params['const'] 
            + res_mod.params['device_before_sleep'] * screen_range
            + res_mod.params['physical_activity'] * act_lvl
            + res_mod.params['screen_x_activity'] * screen_range * act_lvl)
    ax.plot(screen_range, pred, marker='o', label=label, color=color, linewidth=2)

ax.set_xlabel('Device Use Before Sleep')
ax.set_ylabel('Predicted Difficulty Falling Asleep')
ax.set_title('Interaction: Screen Time x Physical Activity on Sleep Difficulty')
ax.legend()
plt.tight_layout()
plt.show()
```

The interaction term is not significant, so exercise does not detectably buffer the effect of screen time on sleep difficulty. The predicted values for low-activity and high-activity students run nearly parallel across screen time levels, meaning the screen-sleep slope is the same regardless of how much someone exercises. The buffering hypothesis does not hold in these data. One caveat is that the survey measures exercise frequency, not timing. If the buffering effect depends on working out in the evening or close to bedtime, a general frequency measure would miss it.

#### Analysis 5: Model Comparison

A single ordinal logit doesn't tell us whether these relationships are best captured linearly or if there's something more complex going on. We compare OLS, ordinal logistic regression, and a random forest on the same prediction task: all survey predictors forecasting academic performance. If the random forest substantially outperforms the linear models, that's a sign we're missing nonlinear structure or interactions we haven't specified.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

feature_cols = ['device_before_sleep', 'physical_activity', 'sleep_quality',
                'sleep_hours', 'difficulty_falling_asleep', 'night_awakenings',
                'caffeine_consumption', 'academic_stress', 'gender', 'year_of_study']

model_df = df1[feature_cols + ['academic_performance']].dropna()
X = model_df[feature_cols]
y = model_df['academic_performance']

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Logistic regression (multinomial approximation of ordinal)
lr_scores = cross_val_score(LogisticRegression(max_iter=1000, random_state=42), X_scaled, y, cv=cv, scoring='accuracy')

# Random forest with hyperparameter grid search
rf_params = {'n_estimators': [100, 200], 'max_depth': [3, 5, 10, None], 'min_samples_leaf': [5, 10, 20]}
rf_grid = GridSearchCV(RandomForestClassifier(random_state=42), rf_params, cv=cv, scoring='accuracy', n_jobs=-1)
rf_grid.fit(X_scaled, y)

print(f'Logistic Regression: {lr_scores.mean():.3f} +/- {lr_scores.std():.3f}')
print(f'Random Forest:       {rf_grid.best_score_:.3f} (best params: {rf_grid.best_params_})')

# Feature importance from best RF
importances = rf_grid.best_estimator_.feature_importances_
feat_imp = pd.Series(importances, index=feature_cols).sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(8, 5))
feat_imp.plot(kind='barh', ax=ax, color='#4E79A7', edgecolor='black')
ax.set_xlabel('Feature Importance')
ax.set_title('Random Forest: Feature Importance for Academic Performance')
plt.tight_layout()
plt.show()
```

All three models land around 49-50% accuracy on a five-class outcome, well above the 20% chance baseline but far from strong prediction. The random forest barely edges out logistic regression, which means the linear models are already capturing most of the structure. There is no hidden nonlinear signal that a more flexible model can exploit. Feature importances from the random forest line up with the regression coefficients: sleep-related variables and stress dominate, with screen time contributing modestly.

#### Analysis 6: Permutation Tests

Regression p-values assume well-behaved residuals, which is optimistic with ordinal data. Permutation tests make no distributional assumptions: we shuffle the outcome variable, refit the model, and count how often the permuted coefficient is as extreme as the real one. We run this for the three key predictors across both sleep and academic outcomes.

```python
np.random.seed(42)
n_perm = 5000

perm_tests = [
    ('device_before_sleep', 'difficulty_falling_asleep'),
    ('device_before_sleep', 'academic_performance'),
    ('physical_activity', 'sleep_quality'),
    ('physical_activity', 'academic_performance'),
    ('sleep_hours', 'academic_performance'),
    ('caffeine_consumption', 'difficulty_falling_asleep'),
]

perm_results = []
for predictor, outcome in perm_tests:
    cols = [predictor] + controls
    pdf = df1[cols + [outcome]].dropna()
    X_real = sm.add_constant(pdf[cols].values.astype(float))
    y_real = pdf[outcome].values.astype(float)
    
    # index of predictor in the design matrix (0 = const, 1 = predictor)
    pred_idx = 1
    real_coef = sm.OLS(y_real, X_real).fit(disp=False).params[pred_idx]
    
    perm_coefs = np.empty(n_perm)
    for i in range(n_perm):
        y_shuf = np.random.permutation(y_real)
        perm_coefs[i] = sm.OLS(y_shuf, X_real).fit(disp=False).params[pred_idx]
    
    p_perm = np.mean(np.abs(perm_coefs) >= np.abs(real_coef))
    perm_results.append({
        'predictor': predictor, 'outcome': outcome,
        'coef': real_coef, 'perm_p': p_perm, 'n': len(pdf)
    })

perm_df = pd.DataFrame(perm_results)
print(perm_df.to_string(index=False))

# Null distribution histograms for two key relationships
fig, axes = plt.subplots(1, 2, figsize=(13, 4))
for ax, (pred, out) in zip(axes, [('device_before_sleep', 'difficulty_falling_asleep'), 
                                    ('physical_activity', 'academic_performance')]):
    cols = [pred] + controls
    pdf = df1[cols + [out]].dropna()
    X_real = sm.add_constant(pdf[cols].values.astype(float))
    y_real = pdf[out].values.astype(float)
    real_c = sm.OLS(y_real, X_real).fit(disp=False).params[1]
    
    null_dist = np.empty(n_perm)
    for i in range(n_perm):
        y_shuf = np.random.permutation(y_real)
        null_dist[i] = sm.OLS(y_shuf, X_real).fit(disp=False).params[1]
    
    ax.hist(null_dist, bins=40, color='#B0B0B0', edgecolor='black', alpha=0.7)
    ax.axvline(real_c, color='#E15759', linewidth=2, label=f'observed = {real_c:.4f}')
    ax.set_title(f'{pred.replace("_"," ").title()} -> {out.replace("_"," ").title()}')
    ax.set_xlabel('Coefficient under null')
    ax.legend(fontsize=9)

plt.tight_layout()
plt.show()
```

Permutation tests largely confirm the parametric results. Screen time predicting sleep difficulty and grades both survive easily, as do sleep duration predicting grades and caffeine predicting sleep difficulty. Physical activity predicting grades is the one that falls short of conventional significance (p = .099), placing it in the "real but modest" category. The null distributions show the observed coefficients for the screen-sleep relationship sitting well outside the permuted range, which reinforces that the association is not an artifact of distributional assumptions in the regression.

#### Analysis 7: Comparing Predictor Strengths

With multiple predictors in play, it's worth asking which ones actually move the needle on academic performance and sleep. We pull standardized coefficients from the full survey regression and the CMU OLS model and plot them side by side. This makes it easy to see whether screen time, exercise, caffeine, or stress matters most, and whether the ranking looks similar across datasets.

```python
# Standardized coefficients for survey: all predictors -> academic performance
all_preds = ['device_before_sleep', 'physical_activity', 'sleep_quality', 'sleep_hours',
             'difficulty_falling_asleep', 'night_awakenings', 'caffeine_consumption', 
             'academic_stress', 'gender', 'year_of_study']

sdf = df1[all_preds + ['academic_performance']].dropna()
X_std = (sdf[all_preds] - sdf[all_preds].mean()) / sdf[all_preds].std()
X_std = sm.add_constant(X_std)
y_std = (sdf['academic_performance'] - sdf['academic_performance'].mean()) / sdf['academic_performance'].std()
ols_full = sm.OLS(y_std, X_std).fit()

surv_coefs = ols_full.params.drop('const')
surv_ci = ols_full.conf_int().drop('const')

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Survey coefficient plot
order = surv_coefs.abs().sort_values().index
colors = ['#E15759' if v < 0 else '#4E79A7' for v in surv_coefs[order]]
axes[0].barh(range(len(order)), surv_coefs[order], xerr=[surv_coefs[order] - surv_ci.loc[order, 0], 
             surv_ci.loc[order, 1] - surv_coefs[order]], color=colors, edgecolor='black', capsize=3)
axes[0].set_yticks(range(len(order)))
axes[0].set_yticklabels([n.replace('_', ' ').title() for n in order])
axes[0].axvline(0, color='black', linewidth=0.8)
axes[0].set_xlabel('Standardized Coefficient')
axes[0].set_title('Survey: Predictors of Academic Performance')

# CMU coefficient plot
cmu_preds = ['TotalSleepTime', 'bedtime_mssd', 'midpoint_sleep', 'daytime_sleep', 'demo_gender']
cdf = df2[cmu_preds + ['term_gpa']].dropna()
X_cmu_std = (cdf[cmu_preds] - cdf[cmu_preds].mean()) / cdf[cmu_preds].std()
X_cmu_std = sm.add_constant(X_cmu_std)
y_cmu_std = (cdf['term_gpa'] - cdf['term_gpa'].mean()) / cdf['term_gpa'].std()
ols_cmu_std = sm.OLS(y_cmu_std, X_cmu_std).fit()

cmu_coefs = ols_cmu_std.params.drop('const')
cmu_ci = ols_cmu_std.conf_int().drop('const')
order_c = cmu_coefs.abs().sort_values().index
colors_c = ['#E15759' if v < 0 else '#4E79A7' for v in cmu_coefs[order_c]]
axes[1].barh(range(len(order_c)), cmu_coefs[order_c], xerr=[cmu_coefs[order_c] - cmu_ci.loc[order_c, 0],
             cmu_ci.loc[order_c, 1] - cmu_coefs[order_c]], color=colors_c, edgecolor='black', capsize=3)
axes[1].set_yticks(range(len(order_c)))
axes[1].set_yticklabels([n.replace('_', ' ').title() for n in order_c])
axes[1].axvline(0, color='black', linewidth=0.8)
axes[1].set_xlabel('Standardized Coefficient')
axes[1].set_title('CMU: Predictors of Term GPA')

plt.tight_layout()
plt.show()
```

Standardized coefficients tell a consistent story across datasets. In the survey, stress and caffeine have the largest absolute effects on grades, with sleep duration and difficulty falling asleep close behind. Screen time ranks in the middle, consistent with its role as an upstream variable that works partly through sleep rather than hitting grades directly. In the CMU data, sleep timing and daytime napping are the strongest predictors of GPA, both negative. The ranking differs because the datasets measure different constructs, but the direction is the same: sleep-related variables dominate the prediction of academic outcomes in both.

## Discussion and Conclusion

We set out to answer three questions: does screen time before bed predict worse sleep, does that sleep disruption carry over into grades, and does physical activity buffer any of it? The short answer is yes to the first two and probably not to the third.

Across all three datasets, screen use before bed was consistently tied to worse sleep outcomes. In the survey data, device use was one of the strongest predictors of difficulty falling asleep even after controlling for caffeine, stress, gender, and year of study. The mediation analysis showed that about 14% of screen time's total association with academic performance runs indirectly through sleep difficulty. That indirect effect was statistically significant (bootstrap CI excluded zero), but the modest size means sleep difficulty is only part of the story and other pathways likely matter too.

The sleep-to-grades connection held up across measurement methods. In the survey data, difficulty falling asleep and sleep hours both predicted academic performance in the ordinal logistic model. In the CMU Fitbit data, total sleep time and bedtime regularity both predicted term GPA with objective measurements from US college students. The fact that these associations appeared in both self-reported and device-measured data, from different countries and universities, makes it harder to write off as survey noise. The standardized coefficient plots showed that sleep-related variables consistently ranked among the top predictors of academic outcomes in both datasets.

The physical activity moderation story was less convincing. The interaction term between screen time and exercise was not significant in predicting sleep difficulty, and the EDA stratified regressions showed nearly identical slopes for high-activity and low-activity students. The random forest feature importances did rank physical activity as a relevant predictor of academic performance on its own, but the buffering effect we hypothesized did not materialize. One possible explanation is that the survey measured general exercise frequency, not timing relative to sleep, and prior research suggesting exercise buffers screen time effects focused on stress rather than sleep difficulty specifically. It may also be that the buffering effect exists but is small enough that our sample couldn't detect it, especially given the imbalanced activity groups in the survey data.

The permutation tests confirmed the parametric results. Screen time predicting sleep difficulty and sleep variables predicting academic performance both survived permutation at conventional significance levels. The caffeine and exercise associations with sleep were weaker but directionally consistent, which matches what we'd expect from variables that are real but modest predictors in a noisy system.

The model comparison between logistic regression and a random forest was informative. The random forest did not dramatically outperform logistic regression on academic performance prediction, which suggests the relationships are roughly linear and we're not missing major nonlinear interactions. The feature importance rankings lined up with the regression coefficients, giving us more confidence that the linear models are capturing the right structure.

There are real limitations here. The survey dataset comes from a university in Bangladesh, and cultural differences in screen habits, academic systems, and lifestyle could affect how well these findings generalize to US students. All survey measures are ordinal and self-reported, which introduces both measurement error and potential social desirability bias. The Dartmouth sensor dataset had only 30 students with complete data, which is too small to draw strong conclusions from on its own. And all three datasets are cross-sectional, so we can't make causal claims. Students who sleep poorly might use their phones more because they can't sleep, rather than the other way around. Longitudinal data tracking the same students over time would be needed to untangle that.

If this work continued, the most valuable next step would be collecting longitudinal data with both objective sleep tracking and app-level screen time logs from the same students over an academic term. That would let us test within-person effects: on nights when a given student uses their phone more, do they sleep worse and perform worse the next day? It would also be worth measuring exercise timing rather than just frequency, since the buffering hypothesis might depend on when someone works out relative to when they go to bed. A larger sample with objective activity data would give the moderation analysis the statistical power it lacked here.
